## 1. Input Parameters of the Prototype

- **neighbor_threshold** — maximum distance allowed when connecting two graph nodes
- **offline_threshold** — maximum allowed disconnected flight time

## 2. How Flight Time Is Computed

For every route segment, the flight time is calculated from the Euclidean distance and UAV speed.

### Formula

\[
t = \frac{d}{v}
\]

where:
- \(t\) — travel time
- \(d\) — Euclidean distance between two nodes
- \(v\) — UAV speed

In the code, this is implemented as:

`time_cost = dist / speed`

The total path time is the sum of all segment times along the final feasible route.


```python
# Example of time computation
dist = 22.80
speed = 1.2
time_cost = dist / speed
print(f"Distance = {dist}")
print(f"Speed = {speed}")
print(f"Time = {time_cost:.2f} min")
```

## 4. Coverage Map and Boundary Formation

The environment is built in two stages:

1. **Coverage map generation**  
   Synthetic Wi-Fi access points are distributed in clusters over the navigation area.

2. **Boundary generation**  
   After cluster points are created, a boundary is constructed around each cluster.  
   In this prototype, the boundary is built using the **Convex Hull** method.

The Convex Hull forms the smallest convex polygon enclosing the cluster points.  
This polygon is later used as a reconnection boundary (CR-Zone boundary).


## 5. Pipeline Overview

The prototype follows this workflow:

1. Coverage Map Generation  
2. Boundary Generation
