## Parameters Explanation
- A, B: start and goal points
- num_clusters: number of clusters
- cluster_spread: cluster size
- min_center_dist: minimum distance between clusters
- area_limit: map size
- neighbor_threshold: max edge length
- offline_threshold: max disconnection time
- speed: UAV speed


## Time Calculation
time = distance / speed
Total path time is the sum of all segment times.
